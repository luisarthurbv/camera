# Data Analyzer: PDF Extraction and Analysis (Markdown version)

This notebook extracts text from PDF documents as Markdown (preserving tables and structure) and performs automated analysis/summarization.

## 1. Setup and Imports

```python
from markitdown import MarkItDown
import os
```

## 2. PDF Text Extraction

Define a helper function to extract text as Markdown from a PDF file path.

```python
def extract_markdown_from_pdf(pdf_path):
    if not os.path.exists(pdf_path):
        return f"Error: File {pdf_path} not found."
    
    md = MarkItDown()
    result = md.convert(pdf_path)
    return result.text_content

# Example usage:
resources_dir = 'resources'
pdf_path = os.path.join(resources_dir, 'ata.pdf')
output_path = os.path.join(resources_dir, 'ata.md')

if os.path.exists(pdf_path):
    md_content = extract_markdown_from_pdf(pdf_path)
    with open(output_path, 'w') as f:
        f.write(md_content)
    print(f"Extracted Markdown saved to {output_path}")
    print("Preview:")
    print(md_content[:500])
```
